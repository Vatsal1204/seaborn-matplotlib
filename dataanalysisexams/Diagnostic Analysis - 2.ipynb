```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("dream_tracker_dataset.csv")

print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
print(df.describe(include="all"))

sns.histplot(df["Sleep_Hours"], kde=True, bins=20, color='skyblue')
plt.title("Distribution of Sleep Hours")
plt.xlabel("Sleep Hours")
plt.ylabel("Count")
plt.show()

sns.countplot(x="Dream_Type", data=df, palette="Set2")
plt.title("Dream Type Frequency")
plt.xlabel("Dream Type")
plt.ylabel("Count")
plt.show()

sns.boxplot(x="Dream_Type", y="Stress_Level_Before_Sleep", data=df, palette="Set3")
plt.title("Stress Level Before Sleep by Dream Type")
plt.xlabel("Dream Type")
plt.ylabel("Stress Level")
plt.show()

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
